```python
import numpy as np
import matplotlib.pyplot as plt
```

**1.2** Toss a coin n times and record whether or not proportion of heads is within $0.1$ of $0.5$, i.e. between $0.4$ and $0.6$. Repeat experiment $100$ times. How large must $n$ be so that approximately $95$ out of $100$ times the proportion of heads is between $0.4$ and $0.6$?

```python
N = 100
M = 100
prop_mean=np.zeros(N)
for i in range(N):
  proportion=np.zeros(M)
  proportion_heads=np.zeros(M)
  for j in range(M):
    batch = np.random.randint(low=0, high=2, size=i+1)
    proportion_heads[j]= np.mean(batch)
    if proportion_heads[j] >= 0.4 and proportion_heads[j] <= 0.6:
      proportion[j]=1
    prop_mean[i]=np.mean(proportion)

plt.plot(prop_mean, label='Proportion of Heads')
plt.axhline(y=0.95, linestyle='dotted', color='r')
plt.legend()
plt.show()
```
