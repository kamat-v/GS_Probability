Simulate a racquetball game; Player A has a 0.6 chance of winning a point while serving, 0.5 chance of winning when opponent serving. Points only get added when server wins a point. First to 21 wins. Player A begins serving. Using simulations, estimate the probability that they win. (If you remove the win by 2 condition -- see below -- the probability appears to be $0.82$.)

**Add-on**: Modify the simulation so that the winner has to win by 2 -- as is typical in racquet sports. Probability appears to be $0.8$

```python
N=10000
R=np.zeros(N)
for i in range(N):
  p1=0 #player 1 starts with 0
  p2=0 #player 2 starts with 0
  serve=0 #player 1 is on serve
  while (p1<21 and p2<21) or (p1>=20 and p2>=20 and abs(p1-p2)<2): #adding a win by 2 condition not mentioned in original problem
    if serve==0:
      if np.random.uniform(0,1)<=0.6:
        p1+=1
      else:
        serve=1
    else:
      if np.random.uniform(0,1)<=0.5:
        serve=0
      else:
        p2+=1
  if p1==21:
    R[i]=1
    print(f'Player A wins in {i+1}th game by a score of {p1} -- {p2}')
  elif p2==21:
    R[i]=0
    print(f'Player B wins in {i+1}th game by a score of {p2} -- {p1}')
  else:
    R[i]=0
print(f'Probability that player A wins is {np.mean(R)}')
```
