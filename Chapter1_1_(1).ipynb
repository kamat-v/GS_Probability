**1.1** Toss a coin $n$ times and print out after every $100$ tosses the proportion of heads minus $1/2$. Do these numbers appear to approach $0$ as $n$ *increases*? Modify the program again to print out, every $100$ times, both of the following quantities: the proportion of heads minus $1/2$, and the number of heads minus half the number of tosses. Do these numbers appear to approach $0$ as $n$ increases?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 100
N = 200
pl1 = np.zeros(N)
pl2 = np.zeros(N)
# empty list to store the results of each batch of tosses
X = []
# total number of heads
total_heads = 0

for i in range(N):
    # Generate a new batch of tosses and append it to X
    current_batch = np.random.randint(low=0, high=2, size=M)
    X.append(current_batch)

    # Update the total number of heads by adding the heads from the current batch
    total_heads += np.sum(current_batch)

    # Calculate the total number of tosses so far
    total_tosses = M * (i + 1)

    # Calculate the mean proportion of heads across all tosses so far
    mean_proportion_heads = total_heads / total_tosses

    # Calculate the two quantities to print and store
    proportion_diff = mean_proportion_heads - 0.5
    heads_minus_half_tosses = total_heads - total_tosses / 2

    print(f'After {i+1}th batch of 100 tosses, num. heads minus half is {proportion_diff} and num. of heads minus half the total tosses is {heads_minus_half_tosses}')

    pl1[i] = proportion_diff
    pl2[i] = heads_minus_half_tosses

plt.plot(pl1, label='Proportion of Heads - 1/2')
plt.plot(pl2, label='Heads - Half Tosses')
plt.xlabel('Batch (x100 tosses)')
plt.ylabel('Value')
plt.title('Coin Toss Simulation Results')
plt.legend()
plt.show()
```
