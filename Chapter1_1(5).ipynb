Consider the bet that all three dice will turn up sixes at least once in $n$ rolls of $3$ dice. Calculate $f(n)$, the probability that at least one triple six turn up when $3$ dice are rolled $n$ times. Determine the smallest value of $n$ necessary for a favorable bet that a triple-six will occur when three dice are rolled $n$ times.

```python
N=150 #number of simulations for a favorable bet, depending on DeMoivre's calculations
calc=0
for i in range(N):
  arr=np.random.randint(low=1,high=7,size=3)
  if sum(arr)==18:
    calc+=1
print(calc)
```
